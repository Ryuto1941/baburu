```python
q = int(input('データ数は?\n'))
a = int(input('昇順なら0、降順なら1を入力\n'))

import random
if a == 0:
    def baburusort(data, n):
        for i in range(n):                                     
            for j in range(n-1, i, -1):                        
                if data[j-1] > data[j]:                         
                    data[j], data[j-1] = data[j-1], data[j]     
                print('i=', i, 'j=', j, data)
        return data
elif a == 1:
    def babulusort(data, n):
        for i in range(n):                                     
            for j in range(n-1, i, -1):                        
                if data[j-1] < data[j]:                         
                    data[j], data[j-1] = data[j-1], data[j]     
                print('i=', i, 'j=', j, data)
        return data
else:
    print('エラー、昇順、降順の選択に誤りがあります')

y = random.sample(range(q), k=q)
print('初期値=', y)
y = baburusort(y, len(y))
```
